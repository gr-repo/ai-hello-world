# Graphviz

https://graphviz.org/doc/info/shapes.html

```python
from graphviz import Digraph

dot = Digraph('G', filename='app_llm_graph', format='png')

# Define the outer container cluster
with dot.subgraph(name='cluster_container') as c:
    c.attr(label='Container running in Codespace', style='rounded,filled', color='lightgrey')
    c.node('App', shape='box', style='filled', fillcolor='white')

# LLM node outside the container
dot.node('LLM', shape='cylinder', style='filled', fillcolor='lightblue')

# Edge from App (inside container) to LLM
dot.edge('App', 'LLM', label='calls out to')

# Show inline in notebook
dot
```

```python
dot = Digraph('G', filename='app_llm_graph', format='png')

# Make graph go left to right
dot = Digraph('G', filename='app_llm_graph', format='png')
dot.attr(rankdir='LR')  # Make arrows go left to right

# Define the network container cluster
with dot.subgraph(name='cluster_container1') as d:
    d.attr(label='Corporate network', style='rounded,filled', color='lightgrey')
    d.node('VS Code', shape='box', style='filled', fillcolor='white', label='VS Code: edit files here')

# Define the outer container cluster
with dot.subgraph(name='cluster_container') as c:
    c.attr(label='Codespaces: Container', style='rounded,filled', color='lightgrey')
    c.node('App', shape='box', style='filled', fillcolor='white', label='App: runs here')

# LLM node outside the container
dot.node('LLM', shape='box', style='filled', fillcolor='lightblue')

# Edge from App (inside container) to LLM
dot.edge('App', 'LLM', label='calls out to')

# Edge from VS Code to App
dot.edge('VS Code', 'App', label='sync files over SSH')

# Show inline in notebook
dot
```
