## <a id='toc1_3_'></a>[Google Organic Results API](#toc0_)

```python
search = GoogleSearch({
    "q": "Sanlam Organisation Structure", 
    "location": "South Africa",
    "api_key": settings.SERP_API_KEY
  })
result = search.get_dict()
```

```python
# Show search results

print(result)
print(json.dumps(result, indent=4))
```

Drill down into a finding a given page/search result. 

```python
print(result['organic_results'])
print(result['organic_results'][0]['link'])
print(json.dumps(result['organic_results'], indent=4))
```

Filter out ".pdf" links from the **organic search result**

```python

def is_pdf(filename):
    return filename.lower().endswith(".pdf")

# print(len(print(result['organic_results'])))
len(result['organic_results'])

for res in result['organic_results']:
    if is_pdf(res['link']):
        print("Skipping pdf:", res['link'])
        continue;

def get_first_non_pdf(results):
    for res in results:
        if not is_pdf(res['link']):
            return res
        
    raise Exception("Could not find any non-PDF links")

first_non_pdf_link = get_first_non_pdf(result['organic_results'])
print()
print()
print("First non-PDF link:")
print(json.dumps(first_non_pdf_link, indent=4))
    #raise Exceptio
```

# <a id='toc2_'></a>[Google Latest From API](#toc0_)

Finds news articles on topics. This is handy but forces you to choose a news outlet. 

https://serpapi.com/latest_from

```python
from serpapi import GoogleSearch

params = {
  "q": "bbc",
  "hl": "en",
  "gl": "us",
  "device": "desktop",
  "api_key": settings.SERP_API_KEY
}

search = GoogleSearch(params)
results = search.get_dict()
latest_from = results["latest_from"]
print(json.dumps(latest_from, indent=4))
```
